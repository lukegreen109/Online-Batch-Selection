```python
# Imports and helper utilities for correlation column distribution visualization
import os
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, Dropdown, ToggleButtons
from pathlib import Path

# Ensure plots inline
%matplotlib inline

CORR_DIR = Path('..') / 'correlations'
assert CORR_DIR.exists(), f"Correlation directory not found: {CORR_DIR}"

# Discover correlation files (sorted numerically by the embedded number)
def list_corr_files():
    files = [p for p in CORR_DIR.glob('correlations*.npy')]
    # extract trailing integer after 'correlations'
    def key(p):
        stem = p.stem  # e.g., correlations100
        num = ''.join(ch for ch in stem if ch.isdigit())
        try:
            return int(num)
        except ValueError:
            return 0
    return sorted(files, key=key)

CORR_FILES = list_corr_files()
print(f"Found {len(CORR_FILES)} correlation arrays:")
for f in CORR_FILES:
    print(' -', f.name)

# Caching loaded arrays to avoid repeated disk IO
_loaded = {}

def load_array(path: Path):
    if path not in _loaded:
        arr = np.load(path)
        # Ensure 2D (if 1D, treat as single column)
        if arr.ndim == 1:
            arr = arr[:, None]
        _loaded[path] = arr
    return _loaded[path]


def describe_array(arr: np.ndarray):
    return {
        'shape': arr.shape,
        'min': float(np.min(arr)),
        'max': float(np.max(arr)),
        'mean': float(np.mean(arr)),
        'std': float(np.std(arr)),
    }

print("\nUsage: Use the per-file cells below to interactively explore column distributions.")
```

```python
# Helper plotting function (robust to missing scipy)
try:
    from scipy.stats import gaussian_kde  # Optional
    _HAVE_SCIPY = True
except Exception:
    gaussian_kde = None
    _HAVE_SCIPY = False


def plot_column_distribution(arr: np.ndarray, col: int, bins: int = 50, kde: bool = True):
    col_data = arr[:, col]
    plt.figure(figsize=(6,4))
    plt.hist(col_data, bins=bins, density=True, alpha=0.55, color='#1f77b4', edgecolor='white', label='Histogram')
    if kde and _HAVE_SCIPY and gaussian_kde is not None and len(np.unique(col_data)) > 1:
        try:
            kde_est = gaussian_kde(col_data)
            xs = np.linspace(col_data.min(), col_data.max(), 256)
            plt.plot(xs, kde_est(xs), color='crimson', linewidth=2, label='KDE')
        except Exception as e:
            print(f"KDE failed: {e}")
    elif kde and not _HAVE_SCIPY:
        print("scipy not installed: skipping KDE (pip install scipy to enable)")
    plt.title(f'Column {col} distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```

```python
# correlations25.npy interactive exploration
file_path = [p for p in CORR_FILES if p.name == 'correlations25.npy'][0]
arr = load_array(file_path)
print(file_path.name, describe_array(arr))

@interact(
    column=IntSlider(min=0, max=load_array(file_path).shape[1]-1, step=1, value=0, description='Column'),
    bins=IntSlider(min=10, max=150, step=5, value=50, description='Bins'),
    kde=ToggleButtons(options=[('KDE On', True), ('KDE Off', False)], description='KDE')
)
def _view(column, bins, kde):
    plot_column_distribution(arr, column, bins=bins, kde=kde)
```
