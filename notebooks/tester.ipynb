```python
import math
```

```python
def should_log_snapshot(step: int, total_steps: int) -> bool:
    """Decide whether to log a snapshot at this step.

    This is a deterministic, state-free schedule that depends only on the
    current (0-indexed) step and the total number of training steps.

    It logs frequently early in training and increasingly sparsely later by
    spacing snapshot "boundaries" uniformly in log(step) (a common principle:
    constant resolution per unit of log-time).

    Conventions:
    - step ranges from 0 .. total_steps (inclusive), where total_steps is the
      number of optimizer updates that will be taken.
    - Always logs at step 0 and at step == total_steps.
    """
    if total_steps < 0:
        raise ValueError(f"total_steps must be >= 0, got {total_steps}")
    if step <= 0:
        return True
    if step >= total_steps:
        return True
    if total_steps == 0:
        return False

    # Target number of snapshots grows slowly with run length.
    # For example: total_steps≈1e5 -> ~85 buckets.
    num_buckets = max(1, int(math.ceil(math.log2(total_steps + 1))) * 5)
    denom = math.log1p(total_steps)
    bucket = int(math.floor(num_buckets * math.log1p(step) / denom))
    prev_bucket = int(math.floor(num_buckets * math.log1p(step - 1) / denom))
    return bucket != prev_bucket
```

```python
n = 152 * 200
count = 0
for step in range(n):
    if should_log_snapshot(step,n):
        print(f"{step}/{n}")
        count += 1
print(count)
```
