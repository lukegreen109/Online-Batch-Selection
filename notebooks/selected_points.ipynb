```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
selected_points = np.load('/home/toddc/research/Online-Batch-Selection/selected_points/MNIST_Noise/{"bsel":"DivBS","seed":1,"model":"LeNet","opt":"adamw","bs":320,"ratio":0.1,"lr":0.001,"wd":0.01}.npy')
```

```python
num_epochs, num_data = selected_points.shape
```

```python
num_selected = selected_points.sum(axis=0)
```

```python

plt.hist(num_selected, bins=num_epochs)

plt.xlabel('number of times selected')
plt.title('Noisy DivBS MNIST LeNet 100 Epochs')
plt.show()
```

```python
last_selection = np.zeros(num_data)
for i in range(num_epochs)[::-1]:

    curr_epoch = selected_points[i,:]
    mask = (curr_epoch == 1) & (last_selection == 0)
    last_selection[mask] += i
```

```python
plt.hist(last_selection, bins=num_epochs)

plt.xlabel('Last Epoch Selected')
plt.title('RHO-Loss MNIST LeNet 100 Epochs')
plt.show()
```
