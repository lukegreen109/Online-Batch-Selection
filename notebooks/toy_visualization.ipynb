```python
def plot_adaboost_stages(clf, X, y, stages=None, grid_res=200, figsize=(12,8), cmap='coolwarm'):
    """
    Visualize decision surfaces for selected AdaBoost stages.
    - clf: fitted sklearn.AdaBoostClassifier
    - X, y: data (X shape (n,2))
    - stages: list of 1-based stage indices to plot (default: a few evenly spaced including last)
    """

    if not hasattr(clf, "staged_predict_proba"):
        raise ValueError("Classifier has no staged_predict_proba (use sklearn AdaBoostClassifier)")

    X = np.asarray(X)
    y = np.asarray(y).ravel()
    if X.shape[1] != 2:
        raise ValueError("This helper expects 2D inputs (n,2) for plotting")

    # collect staged probs on training data
    staged_probs = list(clf.staged_predict_proba(X))  # list length = n_estimators
    n_stages = len(staged_probs)
    if n_stages == 0:
        raise ValueError("No staged predictions found (did you fit the classifier?)")

    # choose stages to display
    if stages is None:
        picks = np.unique(np.concatenate((
            np.linspace(1, min(10, n_stages), min(5, n_stages), dtype=int),
            np.linspace(1, n_stages, 5, dtype=int)
        )))
        stages = np.sort(picks).tolist()
    stages = [s for s in stages if 1 <= s <= n_stages]

    # compute per-stage accuracy on training data
    accs = []
    for probs in staged_protraining_opt:
bs:
        preds = np.argmax(probs, axis=1)
        accs.append(accuracy_score(y, preds))

    # build grid
    x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
    y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, grid_res),
                         np.linspace(y_min, y_max, grid_res))
    grid = np.c_[xx.ravel(), yy.ravel()]

    # get staged probs on grid (generator usage requires calling staged_predict_proba(grid))
    staged_probs_grid = list(clf.staged_predict_proba(grid))  # same length as n_stages

    # plot grid of decision surfaces
    n_plots = len(stages)
    ncols = min(4, n_plots)
    nrows = int(np.ceil(n_plots / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize, squeeze=False)
    axes = axes.flatten()

    for ax_idx, stage in enumerate(stages):
        ax = axes[ax_idx]
        probs_grid = staged_probs_grid[stage - 1]  # 1-based stage index
        # if binary, use probability of positive class for smooth contour; else use argmax
        if probs_grid.shape[1] == 2:
            Z = probs_grid[:, 1].reshape(xx.shape)
            cf = ax.contourf(xx, yy, Z, levels=20, cmap=cmap, alpha=0.4)
            ax.contour(xx, yy, Z, levels=[0.5], colors='k', linewidths=1)
        else:
            Z = np.argmax(probs_grid, axis=1).reshape(xx.shape)
            cf = ax.contourf(xx, yy, Z, levels=np.arange(probs_grid.shape[1]+1)-0.5, cmap='tab10', alpha=0.4)

        # scatter training points
        ax.scatter(X[y==0,0], X[y==0,1], c='C0', edgecolor='k', s=30, label='class 0')
        ax.scatter(X[y==1,0], X[y==1,1], c='C1', edgecolor='k', s=30, label='class 1')

        # title with accuracy
        ax.set_title(f"Stage {stage} — acc {accs[stage-1]:.3f}")
        ax.set_xlim(x_min, x_max); ax.set_ylim(y_min, y_max)
        ax.set_xticks([]); ax.set_yticks([])
        ax.legend(loc='best', fontsize='small')

    # hide unused axes
    for j in range(n_plots, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    return fig, axes
```

```python
def plot_adaboost_stages(clf, X, y, stages=None, grid_res=200, figsize=(12,8), cmap='coolwarm',
                         size_by_weight=True, size_range=(20, 180)):
    """
    Visualize decision surfaces for selected AdaBoost stages.
    If clf.sample_weight_history_ exists and size_by_weight=True, point sizes reflect sample weights.
    """
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.metrics import accuracy_score

    if not hasattr(clf, "staged_predict_proba"):
        raise ValueError("Classifier has no staged_predict_proba (use sklearn AdaBoostClassifier)")

    X = np.asarray(X); y = np.asarray(y).ravel()
    if X.shape[1] != 2:
        raise ValueError("This helper expects 2D inputs (n,2) for plotting")

    staged_probs = list(clf.staged_predict_proba(X))
    n_stages = len(staged_probs)
    if n_stages == 0:
        raise ValueError("No staged predictions found (did you fit the classifier?)")

    # choose stages
    if stages is None:
        picks = np.unique(np.concatenate((
            np.linspace(1, min(10, n_stages), min(5, n_stages), dtype=int),
            np.linspace(1, n_stages, 5, dtype=int)
        )))
        stages = np.sort(picks).tolist()
    stages = [s for s in stages if 1 <= s <= n_stages]

    # per-stage accuracy
    accs = []
    for probs in staged_protraining_opt:
bs:
        preds = np.argmax(probs, axis=1)
        accs.append(accuracy_score(y, preds))

    # grid
    x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
    y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, grid_res),
                         np.linspace(y_min, y_max, grid_res))
    grid = np.c_[xx.ravel(), yy.ravel()]
    staged_probs_grid = list(clf.staged_predict_proba(grid))

    # helper to map weights -> sizes
    def weights_to_sizes(w, rng=size_range):
        if w is None:
            return np.full(X.shape[0], (rng[0] + rng[1]) / 2.0, dtype=float)
        w = np.asarray(w, dtype=float)
        if w.ndim != 1 or w.shape[0] != X.shape[0]:
            return np.full(X.shape[0], (rng[0] + rng[1]) / 2.0, dtype=float)
        w = w / (w.max() + 1e-12)          # [0,1]
        w = np.sqrt(w)                     # area-perception friendly
        return rng[0] + (rng[1] - rng[0]) * w

    weight_history = getattr(clf, "sample_weight_history_", None)

    n_plots = len(stages)
    ncols = min(4, n_plots)
    nrows = int(np.ceil(n_plots / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize, squeeze=False)
    axes = axes.flatten()

    for ax_idx, stage in enumerate(stages):
        ax = axes[ax_idx]
        probs_grid = staged_probs_grid[stage - 1]
        if probs_grid.shape[1] == 2:
            Z = probs_grid[:, 1].reshape(xx.shape)
            ax.contourf(xx, yy, Z, levels=20, cmap=cmap, alpha=0.4)
            ax.contour(xx, yy, Z, levels=[0.5], colors='k', linewidths=1)
        else:
            Z = np.argmax(probs_grid, axis=1).reshape(xx.shape)
            ax.contourf(xx, yy, Z, levels=np.arange(probs_grid.shape[1]+1)-0.5, cmap='tab10', alpha=0.4)

        # sizes from weight history (weights used to fit this stage)
        sizes = None
        if size_by_weight and weight_history is not None and len(weight_history) >= stage:
            sizes = weights_to_sizes(weight_history[stage - 1])
        else:
            sizes = weights_to_sizes(None)

        # scatter
        ax.scatter(X[y==0,0], X[y==0,1], c='C0', edgecolor='k', s=sizes[y==0], label='class 0', alpha=0.9)
        ax.scatter(X[y==1,0], X[y==1,1], c='C1', edgecolor='k', s=sizes[y==1], label='class 1', alpha=0.9)

        ax.set_title(f"Stage {stage} — acc {accs[stage-1]:.3f}")
        ax.set_xlim(x_min, x_max); ax.set_ylim(y_min, y_max)
        ax.set_xticks([]); ax.set_yticks([])
        ax.legend(loc='best', fontsize='small')

    for j in range(n_plots, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    return fig, axes
```

```python
def teacher_boost(self, iboost, X, y, sample_weight, random_state):
        """Implement a single boost.

        Perform a single boost according to the discrete SAMME algorithm and return the
        updated sample weights.

        Parameters
        ----------
        iboost : int
            The index of the current boost iteration.

        X : {array-like, sparse matrix} of shape (n_samples, n_features)
            The training input samples.

        y : array-like of shape (n_samples,)
            The target values (class labels).

        sample_weight : array-like of shape (n_samples,)
            The current sample weights.

        random_state : RandomState instance
            The RandomState instance used if the base estimator accepts a
            `random_state` attribute.

        Returns
        -------
        sample_weight : array-like of shape (n_samples,) or None
            The reweighted sample weights.
            If None then boosting has terminated early.

        estimator_weight : float
            The weight for the current boost.
            If None then boosting has terminated early.

        estimator_error : float
            The classification error for the current boost.
            If None then boosting has terminated early.
        """

        if iboost == 0:
            self.sample_weight_history_ = []

        self.sample_weight_history_.append((sample_weight / np.sum(sample_weight)).copy())

        estimator = self._make_estimator(random_state=random_state)

        estimator.fit(X, y, sample_weight=sample_weight)

        y_predict = estimator.predict(X)

        if iboost == 0:
            self.classes_ = getattr(estimator, "classes_", None)
            self.n_classes_ = len(self.classes_)

        # Instances incorrectly classified
        incorrect = y_predict != y

        # Error fraction
        estimator_error = np.mean(np.average(incorrect, weights=sample_weight, axis=0))

        # Stop if classification is perfect
        if estimator_error <= 0:
            return sample_weight, 1.0, 0.0

        n_classes = self.n_classes_

        # Stop if the error is at least as bad as random guessing
        if estimator_error >= 1.0 - (1.0 / n_classes):
            self.estimators_.pop(-1)
            if len(self.estimators_) == 0:
                raise ValueError(
                    "BaseClassifier in AdaBoostClassifier "
                    "ensemble is worse than random, ensemble "
                    "can not be fit."
                )
            return None, None, None

        # Boost weight using multi-class AdaBoost SAMME alg
        estimator_weight = self.learning_rate * (
            np.log((1.0 - estimator_error) / estimator_error) + np.log(n_classes - 1.0)
        )

        # Only boost the weights if it will fit again
        if not iboost == self.n_estimators - 1:
            # Only boost positive weights
            sample_weight = np.exp(
                np.log(sample_weight)
                + estimator_weight * incorrect * (sample_weight > 0)
            )

        return sample_weight, estimator_weight, estimator_error
```

```python
teacher_clf = AdaBoostClassifier(n_estimators=5, random_state=7)
teacher_clf._boost = teacher_boost.__get__(teacher_clf, AdaBoostClassifier)
teacher_clf.fit(X, y)
fig, axes = plot_adaboost_stages(teacher_clf, X, y, stages=[3,4,5])
```

```python
def my_boost(self, iboost, X, y, sample_weight, random_state):
    estimator = self._make_estimator(random_state=random_state)

    if iboost == 0:
        self.sample_weight_history_ = []

    self.sample_weight_history_.append((sample_weight / np.sum(sample_weight)).copy())

    estimator.fit(X, y, sample_weight=sample_weight)

    y_predict = estimator.predict(X)

    if iboost == 0:
        self.classes_ = getattr(estimator, "classes_", None)
        self.n_classes_ = len(self.classes_)

    # Instances incorrectly classified
    incorrect = y_predict != y

    # Error fraction
    estimator_error = np.mean(np.average(incorrect, weights=sample_weight, axis=0))

    # Stop if classification is perfect
    if estimator_error <= 0:
        return sample_weight, 1.0, 0.0

    n_classes = self.n_classes_

    # Stop if the error is at least as bad as random guessing
    if estimator_error >= 1.0 - (1.0 / n_classes):
        self.estimators_.pop(-1)
        if len(self.estimators_) == 0:
            raise ValueError(
                "BaseClassifier in AdaBoostClassifier "
                "ensemble is worse than random, ensemble "
                "can not be fit."
            )
        return None, None, None

    # Boost weight using multi-class AdaBoost SAMME alg
    estimator_weight = self.learning_rate * (
        np.log((1.0 - estimator_error) / estimator_error) + np.log(n_classes - 1.0)
    )

    learnable = (self.teacher.predict(X) == y)
    learnable_but_not_yet_learned = incorrect & learnable

    # Only boost the weights if it will fit again
    if not iboost == self.n_estimators - 1:
        # Only boost positive weights
        sample_weight = np.exp(
            np.log(sample_weight)
            + estimator_weight * learnable_but_not_yet_learned * (sample_weight > 0)
        )

    return sample_weight, estimator_weight, estimator_error
```
